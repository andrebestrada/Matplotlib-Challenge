```python
# Dependencies and Setup
import matplotlib.pyplot as plt
from scipy.stats import linregress
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
mouse_study=pd.merge(mouse_metadata,study_results,how="left",on="Mouse ID")
# Display the data table for preview
mouse_study.head()
```

```python
# Checking the number of mice.
mice_number=len(mouse_study["Mouse ID"].unique())
mice_number
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mouse=mouse_study.loc[mouse_study.duplicated(subset=["Mouse ID", "Timepoint"]), "Mouse ID"].unique()
print(duplicate_mouse)
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate=mouse_study[mouse_study["Mouse ID"]== 'g989'] 
duplicate
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
mouse_study_df=mouse_study[mouse_study["Mouse ID"].isin(duplicate_mouse) == False]
mouse_study_df
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for 
# each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

summary_stat=mouse_study_df.groupby(["Drug Regimen"])
summary_df=pd.DataFrame({"Mean tumor volume":summary_stat["Tumor Volume (mm3)"].mean(),
                         "Median tumor volume":summary_stat["Tumor Volume (mm3)"].median(),
                         "Variance tumor volume":summary_stat["Tumor Volume (mm3)"].var(),
                         "STD tumor volume":summary_stat["Tumor Volume (mm3)"].std(),
                         "SEM tumor volume":summary_stat["Tumor Volume (mm3)"].sem(),
                        })
summary_df.round(2)
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
mean=mouse_study_df[['Drug Regimen','Tumor Volume (mm3)']].groupby(['Drug Regimen']).agg('mean').round(2)
mean=mean.rename(columns = {'Tumor Volume (mm3)':'Mean TV'})

median=mouse_study_df[['Drug Regimen','Tumor Volume (mm3)']].groupby(['Drug Regimen']).agg('median').round(2)
median=median.rename(columns = {'Tumor Volume (mm3)':'Median TV'})

var=mouse_study_df[['Drug Regimen','Tumor Volume (mm3)']].groupby(['Drug Regimen']).agg('var').round(2)
var=var.rename(columns = {'Tumor Volume (mm3)':'Var TV'})

std=mouse_study_df[['Drug Regimen','Tumor Volume (mm3)']].groupby(['Drug Regimen']).agg('std').round(2)
std=std.rename(columns = {'Tumor Volume (mm3)':'Std TV'})

sem=mouse_study_df[['Drug Regimen','Tumor Volume (mm3)']].groupby(['Drug Regimen']).agg('sem').round(2)
sem=sem.rename(columns = {'Tumor Volume (mm3)':'SEM TV'})

new_summary=pd.concat([mean, median,var,std,sem], axis=1, join='inner')
new_summary
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
treatment_df = mouse_study_df[(mouse_study_df["Drug Regimen"] == 'Capomulin') |
                              (mouse_study_df["Drug Regimen"] == 'Ramicane') |
                              (mouse_study_df["Drug Regimen"] == 'Infubinol') |
                              (mouse_study_df["Drug Regimen"] == 'Ceftamin')]

# Start by getting the last (greatest) timepoint for each mouse
max_df=pd.DataFrame(treatment_df.groupby("Mouse ID")["Timepoint"].max()).reset_index()


# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
merge_df=max_df.merge(mouse_study_df, on=["Mouse ID","Timepoint"], how='left').round(2)
merge_df
```

```python
# Put treatments into a list for loop (and later for plot labels)
treatments = ['Capomulin','Ceftamin','Infubinol','Ramicane']

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data=[]

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
# Locate the rows which contain mice on each drug and get the tumor volumes
for treat in treatments:
    treat_df=merge_df[merge_df['Drug Regimen']==treat]
    
    quartiles = treat_df["Tumor Volume (mm3)"].quantile([.25,.5,.75])
    lowerqu = quartiles[0.25].round(2)
    upperqu = quartiles[0.75].round(2)
    iqr = (upperqu - lowerqu).round(2)
    
    print(f"Quartiles analisys for {treat}")
    print(f"The lower quartile is: {lowerqu}")
    print(f"The upper quartile is: {upperqu}")
    print(f"The interquartile range is: {iqr}")
    print(f"The median is: {quartiles[0.5]}")

    # Determine outliers using upper and lower bounds
    lower_bound = (lowerqu - (1.5*iqr)).round(2)
    upper_bound = (upperqu + (1.5*iqr)).round(2)

    print(f"Values below {lower_bound} could be outliers.")
    print(f"Values above {upper_bound} could be outliers.")
    print("------------------")
    
#     Here I append the drug and the tumor volume for each treatment
    tumor_vol_data.append(merge_df[["Drug Regimen","Tumor Volume (mm3)"]])
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest (individually)
capomulin_df=merge_df[merge_df['Drug Regimen']=='Capomulin']
ceftamin_df=merge_df[merge_df['Drug Regimen']=='Ceftamin']
infubinol_df=merge_df[merge_df['Drug Regimen']=='Infubinol']
ramicane_df=merge_df[merge_df['Drug Regimen']=='Ramicane']

fig1, ax1 = plt.subplots(2,2)
green_diamond = dict(markerfacecolor='g', marker='D')

ax1[0, 0].boxplot(capomulin_df["Tumor Volume (mm3)"], flierprops=green_diamond)
ax1[0, 0].set_title('Capomulin Treatment')
ax1[0, 1].boxplot(ceftamin_df["Tumor Volume (mm3)"], flierprops=green_diamond)
ax1[0, 1].set_title('Ceftamin Treatment')
ax1[1, 0].boxplot(infubinol_df["Tumor Volume (mm3)"], flierprops=green_diamond)
ax1[1, 0].set_title('Infubinol Treatment')
ax1[1, 1].boxplot(ramicane_df["Tumor Volume (mm3)"], flierprops=green_diamond)
ax1[1, 1].set_title('Ramicane Treatment')

for ax in ax1.flat:
    ax.set(ylabel='Tumor Volume (mm3)')
    
for ax in ax1.flat:
    ax.label_outer()

plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
# Mouse ID used to visualize scatter plot is g288

capomulin_df=mouse_study_df[mouse_study_df["Mouse ID"]=='g288'].groupby(['Timepoint']).mean()
capomulin_df=capomulin_df['Tumor Volume (mm3)']
capomulin_df.plot()

plt.ylabel("Tumor Volume (mm3)")
plt.xlabel("Time point (days)")
plt.title("Tumor volume vs. Time point for a mouse treated with Capomulin")
plt.xticks(rotation=0)

plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen

# For the scatter plot that shows relation between weight and tumor volume, I used the total population
# that has Capomulin as the treatment. So we have a more accurate relation between the averages of all mouses
# that were treatened with Capomulin.

capomulin_df=mouse_study_df[mouse_study_df["Drug Regimen"]=='Capomulin'].groupby(['Weight (g)']).mean()
capomulin_df.reset_index(inplace=True)

capomulin_df.plot.scatter(x='Weight (g)',y='Tumor Volume (mm3)')
plt.title("Tumor volume vs. Mouse Weight treated with Capomulin")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
capomulin_df=mouse_study_df[mouse_study_df["Drug Regimen"]=='Capomulin'].groupby(['Weight (g)']).mean()
capomulin_df.reset_index(inplace=True)

weight = capomulin_df['Weight (g)']
volume = capomulin_df['Tumor Volume (mm3)']
correlation = st.pearsonr(weight,volume)[0].round(2)
print(f'The correlation coefficient between mouse weight and average tumor volume is {correlation}')

# Define Values for Regression
x_values = capomulin_df['Weight (g)']
y_values = capomulin_df['Tumor Volume (mm3)']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)

# Creating regrees formula
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(16,44),fontsize=15,color="red")

# Title and Labels
plt.title("Tumor volume vs. Mouse Weight treated with Capomulin")
plt.xlabel('Weight')
plt.ylabel('Tumor Volume (mm3)')
print(f"The r-squared is: {(rvalue**2).round(3)}")
plt.show()

# We reproduce the first scatter plot below or regression model
capomulin_df.plot.scatter(x='Weight (g)',y='Tumor Volume (mm3)')
plt.title("Tumor volume vs. Mouse Weight treated with Capomulin")
plt.show()
```
